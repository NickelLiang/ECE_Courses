## 1.1 Functions and Tests
In this section we first read a list of images from a given path, then concatenate these images with a given order.

```python
def read_imgs_from_path(path):
    # Read file names from given path
    files = os.listdir(path)
    # Specify the extension we want to focus on
    extension = '.png'
    # Feedback to make sure we have correct path and file numbers
    print("Directory '" + path + "' has " + str(len(files)) + " files.")
    # Read images and put into a python list. Not using np.array because images are different size
    # Using matplotlib will read .png images as float, causing trouble in problem 3 ZNCC
    # imgs = [plt.imread(path + f) for f in files if f.endswith(extension)]
    # Using PIL to read .png can read images as uint8
    imgs = [np.asarray(Image.open(path + f)) for f in files if f.endswith(extension)]
    return imgs
```

```python
# Provide a path to shreadded images
img_path = './shredded-images/simple_almastatue/'
# Read images as a list of ndarrray
imgs = read_imgs_from_path(img_path)
```

```python
def combine(imgs, order = None):
    # Create an 'order' array if none provided
    if order is None:
        order = np.arange(len(imgs))
    # Sort imgs base on 'order' array
    sorted_imgs = [x for _,x in sorted(zip(order,imgs))]
    # Start with first image
    combined_img = sorted_imgs[0]
    # Combine rest of the images
    for img in sorted_imgs[1:]:
        combined_img = np.concatenate((combined_img, img), axis=1) 
    return combined_img
```

# 2 Re-order
As you noticed, without correctly ordering the
strips, the images didn't quite look correct. Our goal now is to identify the
correct ordering for these strips. We will adopt a very simple greedy
algorithm. For each pair of strips, we will compute how well does strip 1 go
immediately left of strip 2. We will do this my measuring the similarity
between the last column of pixels in strip 1 and the first column of pixels in
strip 2. You can measure similarity using (negative of) the *sum of squared
differences* which is simply the L2 norm of the pixel difference.

Once you have computed the similarity between all pairs of strips, you can
follow a greedy algorithm to composite the image together. Start with a
strip and place the strip that is most compatible with it on the left and
the right side, and so on. Compose the strips together into an image using
this simple greedy algorithm.

## 2.1 Functions and Tests

```python
def l2_norm(col_l, col_r):
    col_l = col_l.astype(np.int16)
    col_r = col_r.astype(np.int16)
    return np.linalg.norm(col_l-col_r)
```

```python
def reorder(imgs):
    # Process until only one element left in imgs
    while(len(imgs) > 1):
        # Initialize vars: min_val is the minimum difference value, imgs[min_l] and imgs[min_r] has least L2-norm among all possible pairs
        min_val, min_l, min_r = np.inf, 0, 0
        # Go over all images left edge
        for l in range(len(imgs)):
            # Go over all images right edge
            for r in range(len(imgs)):
                # Make sure we are not comparing same image
                if (l != r):
                    # Compute L2 norm between left image's right edge and right image's left edge
                    diff = l2_norm(imgs[l][:,-1], imgs[r][:,0])
                    # If difference is smaller than we currently have, update
                    if (diff < min_val):
                        min_val, min_l, min_r = diff, l, r
        # Set left image to be the concatenated image, then delete right image
        imgs[min_l] = np.concatenate((imgs[min_l], imgs[min_r]), axis=1)
        imgs.pop(min_r)
    # Return the only element in imgs list
    return imgs[0]
```

## 2.2 Solution

```python
# Process simple_almastatue
img_path = './shredded-images/simple_almastatue/'
imgs = read_imgs_from_path(img_path)
reordered = reorder(imgs)
plt.imshow(reordered)
plt.show()
plt.imsave('q2_almastatue.png', reordered)

# Process simple_larry-roberts
img_path = './shredded-images/simple_larry-roberts/'
imgs = read_imgs_from_path(img_path)
reordered = reorder(imgs)
plt.imshow(reordered)
plt.show()
plt.imsave('q2_larry-roberts.png', reordered)
```

# 3 Align and Re-order
Next, we will tackle the case where our
shredder also cut out the edges of some of the strips, so that they are no
longer aligned vertically. These are prefixed with `hard_`. We will now
modify the strip similarity function from the previous part to compensate for
this vertical misalignment.

We will vertically slide one strip relative to the other, in the range of 20% 
of the vertical dimension to find the best alignment, and compute
similarity between the overlapping parts for different slide amounts. We
will use the *maximum* similarity over these different slide amounts, as the
similarity between the two strips. Note that you may need to experiment with
similarity functions other than the sum of squared distances. One common
alternative is *zero mean normalized cross correlation*, which is simply the
dot product between the two pixel vectors after they have been normalized to
have zero mean and unit norm.

```python
# This function find the index of first valid row in a -np.inf padded image
# side = 'l' to check the left side of the image, 'r' to check the right side of the image
def find_first_row(img, side = 'l'):
    if (side == 'l'):
        return np.argmax(img[:,0,0] >= 0)
    elif (side == 'r'):
        return np.argmax(img[:,-1,0] >= 0)
    else:
        assert(False)

# This function find the index of last valid row in a -np.inf padded image
def find_last_row(img, side = 'l'):
    if (side == 'l'):
        return img.shape[0] - np.argmax(np.flip(img[:,0,0]) >= 0) - 1
    elif (side == 'r'):
        return img.shape[0] - np.argmax(np.flip(img[:,-1,0]) >= 0) - 1
    else:
        assert(False)

# Concatenate an array of 'height' to the 't'=top or 'b'=bottom of the 'img'.
# The black section will be filled with -np.inf so that we can recover the original image if necessary.
def fill_black(img, height, side = 't'):
    black = np.full((height, img.shape[1], img.shape[2]), -np.inf)
    if (side == 't'):
        return np.concatenate((black, img), axis = 0)
    elif (side == 'b'):
        return np.concatenate((img, black), axis = 0)
    else:
        assert(False)

# Convert img to a uint8 to display and save properly
def save_img(img, name):
    img = img.astype(np.uint8)
    plt.imsave(name, img)
```

```python
# Convert an image from RGB color space to graysacle image
def rgb2gray(img):
    # Normalize to between 1 and 0
    img = img/255
    # A magic value recommended by internet
    rgb_weights = [0.2989, 0.5870, 0.1140]
    # Convert to grayscale
    return np.dot(img[...,:3], rgb_weights)
```

```python
# Combine to images together with necessary black padding. 'img_l' is the left side image, 'img_r' is right side.
# 'offset'=0 means img_l and img_r's top edge are aligned. 
# 'offset'<0 means img_r is higher up than img_l
# 'offset'>0 means img_l is higher up than img_r
def combine_w_vshift(img_l, img_r, offset):
    # Compute some coordinates. We need coordinates of right edge of left img and left edge of right img
    # because that is the edge two images will be combined together
    # Find the first valid pixel along right edge of left img
    l_top_valid = find_first_row(img_l, 'r')
    # Find the last valid pixel along right edge of left img
    l_bottom_valid = find_last_row(img_l, 'r')
    # Find total height of left image
    l_h = img_l.shape[0]
    # Find top black pixel height of left image, along right edge
    l_h_top_blank = l_top_valid
    # Find valid pixel height of left image, along right edge
    l_h_valid = l_bottom_valid - l_top_valid + 1
    # Find bottom black pixel height of left image, along right edge
    l_h_bottom_blank = l_h - l_bottom_valid - 1
    # Find the first valid pixel along left edge of right img
    r_top_valid = find_first_row(img_r, 'l')
    # Find the last valid pixel along left edge of right img
    r_bottom_valid = find_last_row(img_r, 'l')
    # Find the total height of right img
    r_h = img_r.shape[0]
    # Find top black pixel height along left edge of right img
    r_h_top_blank = r_top_valid
    # Find valid pixel height along left edge of right img
    r_h_valid = r_bottom_valid - r_top_valid + 1
    # Find bottom black pixel height along left edge of right img
    r_h_bottom_blank = r_h - r_bottom_valid - 1
    
    # Some print for debug
    # print("tv\t%d\t%d" % (l_top_valid, r_top_valid))
    # print("bv\t%d\t%d" % (l_bottom_valid, r_bottom_valid))
    # print("h\t%d\t%d" % (l_h,r_h))
    # print("htb\t%d\t%d" % (l_h_top_blank,r_h_top_blank))
    # print("hv\t%d\t%d" % (l_h_valid,r_h_valid))
    # print("hbb\t%d\t%d" % (l_h_bottom_blank,r_h_bottom_blank))
        
    # Pad top side first. Offset is considering left img still and move right img
    r_top_offset = l_top_valid + offset - r_top_valid
    if (r_top_offset < 0):
        # Fill inf in left image
        img_l = fill_black(img_l, abs(r_top_offset), 't')
    else:
        # Fill inf in right image
        img_r = fill_black(img_r, abs(r_top_offset), 't')
    
    # Pad bottom. Offset is considering left img still and move right img
    r_bottom_offset = l_h - (l_top_valid + offset + r_h_valid + r_h_bottom_blank)
    if (r_bottom_offset < 0):
        # Fill inf in left img
        img_l = fill_black(img_l, abs(r_bottom_offset), 'b')
    else:
        # Fill inf in right img
        img_r = fill_black(img_r, abs(r_bottom_offset), 'b')

    # Check and make sure left and right img have same height
    assert(img_l.shape[0] == img_r.shape[0])
    
    # Return concatenated image
    return np.concatenate((img_l, img_r), axis=1)
```

```python
# ZNCC implementation 1: Calculate three channel separately then add together
def zncc(col_l, col_r):
    # Convert to float to prevent underflow in following operation
    col_l = col_l.astype(np.float)
    col_r = col_r.astype(np.float)
    
    # Create empty array
    zn_l = np.empty_like(col_l)    
    zn_r = np.empty_like(col_r)
    
    # Make each channel zero mean and norm 1
    for c in range(zn_l.shape[-1]):
        zn_l[:,c] = (col_l[:,c] - np.mean(col_l[:,c])) / np.linalg.norm(col_l[:,c])
        zn_r[:,c] = (col_r[:,c] - np.mean(col_r[:,c])) / np.linalg.norm(col_r[:,c])

    # Calculate ZNCC
    zncc_r = np.dot(zn_l[:,0], zn_r[:,0].T)
    zncc_g = np.dot(zn_l[:,1], zn_r[:,1].T)
    zncc_b = np.dot(zn_l[:,2], zn_r[:,2].T)

    return zncc_r + zncc_g + zncc_b
```

```python
# ZNCC implementation 2: Calculate three channel together
def zncc2(col_l, col_r):
    # Convert to float to prevent underflow
    col_l = col_l.astype(np.float)
    col_r = col_r.astype(np.float)
    
    # Make three channels zero mean and norm 1
    vec_l = (col_l - np.mean(col_l)) / np.linalg.norm(col_l)
    vec_r = (col_r - np.mean(col_r)) / np.linalg.norm(col_r)

    # Calculate ZNCC
    zncc_r = np.dot(vec_l[:,0], vec_r[:,0].T)
    zncc_g = np.dot(vec_l[:,1], vec_r[:,1].T)
    zncc_b = np.dot(vec_l[:,2], vec_r[:,2].T)

    return zncc_r + zncc_g + zncc_b
```

```python
# ZNCC implementation 3: Convert to grayscale then calculate ZNCC
def zncc_gray(col_l, col_r):
    # Convert to float to prevent underflow
    col_l = col_l.astype(np.float)
    col_r = col_r.astype(np.float)
    
    # Convert to grayscale
    col_l_gray = rgb2gray(col_l)
    col_r_gray = rgb2gray(col_r)
    
    # Make it zero mean and norm 1
    vec_l = (col_l_gray - np.mean(col_l_gray)) / np.linalg.norm(col_l_gray)
    vec_r = (col_r_gray - np.mean(col_r_gray)) / np.linalg.norm(col_r_gray)
    
    return np.dot(vec_l, vec_r.T)
```

```python
# This function find the common adjacent pixels between col_l and col_r, with a given offset.
# 'offset'=0, col_l and col_r are top aligned
# 'offset'<0, col_l is still and col_r is higher than col_l
# 'offset'>0, col_l is still and col_r is lower than col_l
def slide_n_reshape(col_l, col_r, offset):
    # When col_l is higher than col_r
    if (offset > 0):
        # If col_r is taller than col_l w/ offset, col_l will not completely cover col_r
        if (col_l.shape[0]-offset < col_r.shape[0]):
            temp_l = col_l[0+offset:]
            temp_r = col_r[:col_l.shape[0]-offset]
        # Otherwise col_l completely covers the col_r, shared pixels are the height of col_r
        else:
            temp_l = col_l[0+offset:0+offset+col_r.shape[0]]
            temp_r = col_r
    # When col_l is lower than col_r
    else:
        # If col_l istaller than col_r w/ offset, col_r will not completely cover col_l
        if (col_r.shape[0]+offset < col_l.shape[0]):
            temp_l = col_l[:col_r.shape[0]+offset]
            temp_r = col_r[0-offset:]
        # Otherwise col_r completely covers the col_l, shared pixels are the height of col_l
        else:
            temp_l = col_l
            temp_r = col_r[0-offset:0-offset+col_l.shape[0]]
    
    # Make sure result column have the same shape
    assert(temp_l.shape == temp_r.shape)
    return temp_l, temp_r
```

```python
# This function find a minimum ZNCC given left and right image and a sliding range
def find_min_zncc(img_l, img_r, offset_percent = 0.2):
    # Same logic as combine_w_vshift function
    l_top_valid = find_first_row(img_l, 'r')
    l_bottom_valid = find_last_row(img_l, 'r')
    l_h_valid = l_bottom_valid - l_top_valid + 1
    r_top_valid = find_first_row(img_r, 'l')
    r_bottom_valid = find_last_row(img_r, 'l')
    r_h_valid = r_bottom_valid - r_top_valid + 1
    
    # Calculate the maximum offset value base on the taller image
    offset_max = (max(l_h_valid, r_h_valid)*offset_percent).astype(int)
    
    # Initialize max value. A higher ZNCC means higher similarity
    # Should change variable name to max_
    min_zncc, min_offset = -np.inf, -np.inf
    for o in range(-offset_max, offset_max):
        # Get common pixels from right edge of left img and left edge of right img
        col_l, col_r = slide_n_reshape(img_l[l_top_valid:l_bottom_valid,-1], img_r[r_top_valid:r_bottom_valid,0], o)
        # Calculate ZNCC
        zncc_val = zncc2(col_l, col_r)
        # Compare with current min value, and update if necessary
        if (zncc_val > min_zncc):
            min_zncc, min_offset = zncc_val, o
    # Return minimum value and offset
    return min_zncc, min_offset
```

```python
# This function will find merge a imgs array given a offset percentage range.
def align_n_reorder(imgs, offset_percent = 0.2):
    # Process until only one element left in imgs
    while(len(imgs) > 1):
        # Initialize vars: min_val is the minimum difference value, imgs[min_l] and imgs[min_r] has least L2-norm among all possible pairs
        min_val, min_l, min_r, min_offset = -np.inf, 0, 0, 0
        # Go over all images left edge
        for l in range(len(imgs)):
            # Go over all images right edge
            for r in range(len(imgs)):
                # Make sure we are not comparing same image
                if (l != r):
                    # Compute ZNCC between left image's right edge and right image's left edge
                    diff, offset = find_min_zncc(imgs[l], imgs[r], offset_percent)
                    # If difference is smaller than we currently have, update
                    if (diff > min_val):
                        min_val, min_l, min_r, min_offset = diff, l, r, offset
        # Set left image to be the concatenated image, then delete right image
        imgs[min_l] = combine_w_vshift(imgs[min_l], imgs[min_r], min_offset)
        # Debug function to show middle stage
        # temp_1 = imgs[min_l].astype(np.uint8)
        # plt.imshow(temp_1)
        # plt.show()
        imgs.pop(min_r)
    # Return the only element in imgs list
    return imgs[0]
```
